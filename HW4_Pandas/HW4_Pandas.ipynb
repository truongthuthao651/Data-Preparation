```python
# a) Read file
df2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data /covid19_cases.csv')
df2.head()
```

```python
df2.shape
```

```python
# b) Create a date column using the data in the dateRep column & pd.to_datetime() function
df2 = df2.assign(date= pd.to_datetime(df2.dateRep))
df2.head()
```

```python
# c) Set data column as the index & sort the index
df2.set_index('date', inplace=True)
df2.head()
```

```python
df2.sort_index().head()
```

```python
# d) Replace all occurrences of United_States_of_America and United_ Kingdom with USA and UK, respectively. 
#Hint: the replace() method can be run on the dataframe as a whole
df2 = df2.replace('United_States_of_America', 'USA').replace('United_Kingdom', 'UK')
df2.head()
```

```python
# e) Using the countriesAndTerritories column, filter the cleaned COVID-19 cases data down to
# Argentina, Brazil, China, Colombia, India, Italy, Mexico, Peru, Russia, Spain, Turkey, the UK, and the USA
cleaned_df2 = df2[df2.countriesAndTerritories.isin(['Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey', 'UK', 'USA'])]
cleaned_df2.head()
```

```python
# f) Pivot the data so that the index contains the dates, the columns contain the country names, and the values are the case counts (the cases column). 
#Be sure to fill in NaN values with 0
cleaned_df2.reset_index().pivot(index='date', columns='countriesAndTerritories', values='cases').fillna(0)
```
