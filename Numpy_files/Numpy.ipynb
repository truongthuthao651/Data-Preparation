```python
#-------K-Means clustering-------: phân cụm data lớn thành K vùng (vd: phân loại KH, thống kê dữ liệu...)
# Chọn k centroids bất kì.   (xđ k = pp elbow ---> {trung bình cộng of [(k/c từ 1 điểm ts tâm cụm)^2]})        # kmeans_init_centers: khởi tạo centers bđầu
# Tính khoảng cách của mỗi điểm tới k điểm centroids, gán điểm đó vào lớp có centroids gần nhất.               # kmeans_asign_labels: gán nhãn ms cho các điểm khi biết centers
# Update centroids là trung bình các điểm dữ liệu trong class đấy.                                             # kmeans_update_centers: update centers ms
# Quay lại bước 2 cho tới khi các điểm centroids ở lần lặp tới không đổi.                                      # has_converged: check ddkien stop
```

```python
import numpy as np
import matplotlib.pyplot as plt              # visualize data sdung đồ thị
from scipy.spatial.distance import cdist     # hỗ trợ tính khoảng cách
```

```python
means = [[2, 2], [9, 2], [4, 9]]
cov = [[2, 0], [0, 2]]
n_samples = 500      # N
n_cluster = 3        # K
X0 = np.random.multivariate_normal(means[0], cov, n_samples)
X1 = np.random.multivariate_normal(means[1], cov, n_samples)
X2 = np.random.multivariate_normal(means[2], cov, n_samples)
X = np.concatenate((X0, X1, X2), axis = 0)
X
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.plot(X[:, 0], X[:, 1], 'bo', markersize=5)
plt.plot()
plt.show()
```

```python
def kmeans_init_centers(X, n_cluster):
  # random k index beetween 0 and shape(X) without duplicate index.
  # Then return X[index] as cluster
  return X[np.random.choice(X.shape[0], n_cluster, replace=False)]
```

```python
# Hàm này dùng để vẽ dữ liệu lên đồ thị
# Random color chỉ làm việc với k <= 4
# Nếu bạn thay đổi k > 4, hãy sửa lại phần random color nhé
# Chỉ sử dụng trong bài toán này thôi nhé.
def kmeans_visualize(X, centers, labels, n_cluster, title):
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title(title)
  plt_colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
 
  for i in range(n_cluster):
    data = X[labels == i]
    plt.plot(data[:, 0], data[:, 1], plt_colors[i] + '^', markersize = 4, label = 'cluster_' + str(i))
    plt.plot(centers[i][0], centers[i][1],  plt_colors[i+4] + 'o', markersize = 10, label = 'center_' + str(i))
  plt.legend()
  plt.show()
```

```python
def kmeans_predict_labels(X, centers):
  D = cdist(X, centers)
  # return index of the closest center
  return np.argmin(D, axis = 1)
```

```python
def kmeans_update_centers(X, labels, n_cluster):
  centers = np.zeros((n_cluster, X.shape[1]))
  for k in range(n_cluster):
    # collect all points assigned to the k-th cluster 
    Xk = X[labels == k, :]
    # take average
    centers[k,:] = np.mean(Xk, axis = 0)
  return centers
```

```python
def kmeans_has_converged(centers, new_centers):
  # return True if two sets of centers are the same
  return (set([tuple(a) for a in centers]) == 
      set([tuple(a) for a in new_centers]))
```

```python
def kmeans(init_centes, init_labels, X, n_cluster):
  centers = init_centes
  labels = init_labels
  times = 0
  while True:
    labels = kmeans_predict_labels(X, centers)
    kmeans_visualize(X, centers, labels, n_cluster, 'Assigned label for data at time = ' + str(times + 1))
    new_centers = kmeans_update_centers(X, labels, n_cluster)
    if kmeans_has_converged(centers, new_centers):
      break
    centers = new_centers
    kmeans_visualize(X, centers, labels, n_cluster, 'Update center possition at time = ' + str(times + 1))
    times += 1
  return (centers, labels, times)
```

```python
init_centers = kmeans_init_centers(X, n_cluster)
print(init_centers)
init_labels = np.zeros(X.shape[0])
kmeans_visualize(X, init_centers, init_labels, n_cluster, 'Init centers in the first run. Assigned all data as cluster 0')
centers, labels, times = kmeans(init_centers, init_labels, X, n_cluster)

print('Done! Kmeans has converged after', times, 'times')
```

```python
### Sample:
def kmeans_init_centers(X, K):
  return X[np.random.choice(X.shape[0], K, replace = False)]        # chọn random K centroids

def kmeans_assign_labels(X, centers):
  D = cdist(X, centers)                                             # k/c data ts centers
  return np.argmin(D, axis = 1)                                     # return điểm gần center nhất

def kmeans_update_centers(X, labels, K):
  centers = np.zeros((K, X.shape[1]))
  for k in range(K):
    Xk = X[labels == k, :]                                           # collect all points thuộc cùng K-center vào thành 1 cluster
    centers[k, :] = np.mean(Xk, axis = 0)                            # update centroids là k/c trung bình các điểm data trong class đó
  return centers 

def has_converged(centers, new_centers):
  return (set([tuple(a) for a in centers]) ==                         # return nếu tập có centroid ms & tập cũ là giống nhau
          set([tuple(a) for a in new_centers]))
```

```python
def kmeans(X, K):
    centers = [kmeans_init_centers(X, K)]
    labels = []
    it = 0 
    while True:
        labels.append(kmeans_assign_labels(X, centers[-1]))
        new_centers = kmeans_update_centers(X, labels[-1], K)
        if has_converged(centers[-1], new_centers):
            break
        centers.append(new_centers)
        it += 1
    return (centers, labels, it)
```
